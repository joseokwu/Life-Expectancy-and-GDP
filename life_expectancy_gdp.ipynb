```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head(10))
```

```python
print(df.Country.unique())
```

```python
#Check Chile's average GDP and Life expectancy
```

```python
df_chile = df[df.Country == 'Chile']
print(df_chile)
```

```python
avg_gdp_chile = np.mean(df_chile.GDP)
avg_lifeexpectancy_chile = np.mean(df_chile['Life expectancy at birth (years)'])
print('The average GDP in Chile is ' + str(avg_gdp_chile) + ' USD')
print('The average Life expectancy in Chile is ' + str(round(avg_lifeexpectancy_chile)) + 'years')
```

```python
#Check if life expectancy has increased overtime in Chile
```

```python
sns.lmplot(data = df_chile, x = 'Year', y = 'Life expectancy at birth (years)')
plt.show()
plt.clf()
```

```python
#As you can see from the scatterplot above,Though there was an abnormal decline between the years of 2008 and 20010, there has otherwise been a significant increase over the years in the life expectancy in Chile.
```

```python
#Check if GDP has increased overtime in Chile
```

```python
sns.lmplot(data = df_chile, x = 'Year', y = 'GDP')
plt.show()
plt.clf()
```

```python
#From the scatterplot above, it can be deduced that the there has been a significant increase in Chile's GDP across the years, with a few abnormal declines like between the years 2000 - 2002, 2008-2009 and 2012 - 2015.
```

```python
#Check life expectancy VS GDP of Chile
```

```python
sns.lmplot(data = df_chile, x = 'GDP', y = 'Life expectancy at birth (years)')
plt.show()
plt.clf()
```

```python
#As is obvious from the line plot above, it shows that there is a direct correlation between chile's life expectancy and it's GDP.
#Now, to use pearson's correlation to measure the strength of this linear relationship.
```

```python
corr, p = pearsonr(df_chile['Life expectancy at birth (years)'], df_chile['GDP'])
print(corr)
```

```python
#The correlation being so close to 1, only goes further to prove our initial assesment from the line graph. That there is a very strong correlation between the life expectancy and the GDP of chile.
```
